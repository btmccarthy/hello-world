#### Please perform the following
a. print all the element tags in the tree  
b. print all the child elements in the tree  
c. print all the author's last names  
d. The following will print the last name of the book that weighs 1.5 units  
```
print(root.find('Book[@Weight="1.5"]/Authors/Author/Last_Name').text)
```
Using the above as a reference, print the last name of the author who's residence is Beijing  

````python
from lxml import etree                                   #imports the object "etree", useful for manipualating xml data
root=etree.XML(data_string)                              #formats xml string into an element tree, "root"

# a. print all the element tags in the tree
for element in root.iter():                              #iterates for every element (labels for "branches") in tree by
    print(element.tag)                                   #traversing each branch, reversing one level, then traversing the next

# b. print all the child elements in the tree
for child in root:                                       #iterates only for first levels down from root
    print(child)
    
# c. print all the author's last names
for element_au in root.iter("Author"):                   #iterates for every element name Author
    print(element_au.find("Last_Name").text)             #prints the "leaf"/data for every element name "Last_Name" under Author
    
# d. The following will print the last name of the book that weighs 1.5 units
print(root.find('Book[@Weight="1.5"]/Authors/Author/Last_Name').text)       #traverses the specified branch (with Book wt=1.5)

# Using the above as a reference, print the last name of the author who's residence is Beijing
print(root.find('Book/Authors/Author[@Residence="Beijing"]/Last_Name').text)  #traverses branch with Author Residence=Beijing
````

# 8. Scraping HTML 
In this problem we use the requests library in combination with the BeautifulSoup library to scrape data from the website: https://money.cnn.com/data/markets/.  Take a look at the website to get a sense for its content.  

````python
import requests
from bs4 import BeautifulSoup                              #useful for parsing/manipulating HTML data in python
url="https://money.cnn.com/data/markets/"
response = requests.get(url)
results_page=BeautifulSoup(response.content, 'lxml')       #deserializes html data into an object python can intepret
print(results_page.prettify())                             #formats object into a viewable parse tree
````

The program below uses the .find() function from BeautifulSoup to extract the text from an html element, <span> with a specific class, "stock-price".

````python
price=results_page.find('span',{'class': 'stock-price'}).get_text()
print(price, type(price))
price=float(price)
print(price, type(price))
````
